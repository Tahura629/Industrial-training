**Assignment 2:** Create a dictionary where keys are student names and values are lists of grades. Calculate the average grade for each student.

```python
# Step 1: Create a dictionary with student names as keys and lists of grades as values
students_grades = {
    "Alice": [85, 90, 88, 92],
    "Bob": [78, 74, 80, 76],
    "Charlie": [92, 95, 89, 91],
    "David": [65, 70, 68, 72],
    "Eve": [88, 84, 82, 86]
}
```

**Assignment 4:** Write a regex to find all the hashtags in a string.

```python
import re

def find_hashtags(text):
    # Regex pattern to find hashtags
    pattern = r'#\w+'
    # Find all matches of the pattern
    hashtags = re.findall(pattern, text)
    return hashtags
```

```python
# Example usage
text = "Loving the sunny weather! #sunny #weather #happyDay #Python3"
hashtags = find_hashtags(text)
print("Hashtags found:", hashtags)
```

**Assignments 5:** Write a regex to find the Bangladesh phone number with all variations

```python
import re

def find_bd_phone_numbers(text):
    # Regex pattern for Bangladeshi phone numbers
    pattern = r'(\+880|880|0)(1[3-9])\d{8}\b'
    # Find all matches of the pattern
    phone_numbers = re.findall(pattern, text)
    return ["".join(number) for number in phone_numbers]
```

```python
# Example usage
text = """
Here are some Bangladeshi phone numbers:
+8801712345678, 8801912345678, 01712345678, 01898765432, 01476543210.
Invalid numbers: 880112345678, 01912345.
"""
numbers = find_bd_phone_numbers(text)
print("Bangladeshi phone numbers found:", numbers)
```

```python
def calculate_factorial(n):
    try:
        # Check if the input is a valid integer
        if not isinstance(n, int):
            raise ValueError("Input must be an integer.")
```

```python
# Factorial is not defined for negative numbers
        if n < 0:
            raise ValueError("Factorial is not defined for negative numbers.")
```

```python
# Base case for factorial: 0! = 1
        if n == 0:
            return 1
```

```python
 # Calculate factorial using a loop
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i

        return factorial

    except ValueError as e:
        return f"Error: {e}"
```

```python
# Write the list of dictionaries to a JSON file
    with open(json_file, mode='w') as json_out:
        json.dump(data, json_out, indent=4)

    print(f"CSV data has been successfully converted to {json_file}.")
```

**Assignment 2:** Create a log file writer that appends log messages to a file with timestamps.


```python
import logging
from datetime import datetime
```

```python
# Function to configure and create a log file
def setup_logger(log_file):
    # Create a logger
    logger = logging.getLogger()
    logger.setLevel(logging.INFO)
```

```python
# Create a file handler that appends logs to the specified file
    file_handler = logging.FileHandler(log_file)
    file_handler.setLevel(logging.INFO)
```

```python
 # Create a log formatter with timestamp
    formatter = logging.Formatter('%(asctime)s - %(message)s', datefmt='%Y-%m-%d %H:%M:%S')
    file_handler.setFormatter(formatter)
```

```python
# Add the handler to the logger
    logger.addHandler(file_handler)

    return logger
```

```python
# Function to log messages
def log_message(logger, message):
    logger.info(message)
```

```python
# Example usage
log_file = 'app_logs.txt'  # Log file to store log messages
```

```python
# Setup the logger
logger = setup_logger(log_file)
```

```python
# Example log messages
log_message(logger, "Application started.")
log_message(logger, "User logged in.")
log_message(logger, "Error: Unable to load file.")
log_message(logger, "Application finished.")

print(f"Logs are being written to {log_file}.")
```

**Assignment 1:** Write a function that calculates compound interest using the formula A = P(1 + r/n)^(nt).

```python
def calculate_compound_interest(P, r, n, t):
    """
    Function to calculate compound interest.

    Parameters:
    P : float : Principal amount (initial investment)
    r : float : Annual interest rate (decimal form, e.g., 5% as 0.05)
    n : int : Number of times interest is compounded per year
    t : int : Time the money is invested for, in years

    Returns:
    float : Amount after interest
    """
    # Calculate compound interest using the formula
    A = P * (1 + r / n) ** (n * t)

    return A
```

```python
# Example usage
P = 1000  # Principal amount
r = 0.05  # Annual interest rate (5%)
n = 4     # Number of times interest is compounded per year (quarterly)
t = 5     # Time in years
```

```python
# Calculate the compound interest
amount = calculate_compound_interest(P, r, n, t)
```

```python
# Display the result
print(f"The amount after {t} years is: ${amount:.2f}")
```

**Assignment 2:** Create a script that prints the current time and updates every second.

```python
import time
import os

def print_current_time():
    try:
        while True:
            # Clear the console
            os.system('cls' if os.name == 'nt' else 'clear')

            # Get and print the current time
            current_time = time.strftime("%H:%M:%S", time.localtime())
            print(f"Current Time: {current_time}")

            # Wait for 1 second
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nExiting program.")

if __name__ == "__main__":
    print_current_time()
```
